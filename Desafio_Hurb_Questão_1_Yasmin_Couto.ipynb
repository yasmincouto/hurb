```python
#Importando pacotes e requisitos para coleta
import requests
import pandas as pd
```

```python
##Espécies dos Pokemons
```

```python
#Requisitando dados das especies e verificando status da resposta
especies = requests.get("https://pokeapi.co/api/v2/pokemon?limit=100000&offset=0")
pokedex_especies = especies.json()

print("Status da requisição:",especies.status_code)
print("Resultado da requisição:",pokedex_especies)
print("Chaves da requisição:",pokedex_especies.keys())
```

```python
#Criando dataframe base df_especies a partir da requisição e selecionando 151 especies de interesse

df_especies = pd.DataFrame.from_dict(pokedex_especies)
df_especies = df_especies['results']
df_especies = df_especies.apply(pd.Series)
df_especies = df_especies[:151]
df_especies = df_especies.rename(columns={'name': 'Pokemon'})

print(df_especies.shape)
df_especies.tail()
```

```python
#Criando função para requisitar habilidades de cada espécie

def get_informacoes(i):
    r = requests.get(i)
    return r
```

```python
#Criando loop para interar entre linhas do dataframe base df_especies
#Requisições serão armazenadas na lista req_informacoes

req_informacoes = []
for i in df_especies['url']:
  x = get_informacoes(i).json()
  req_informacoes.append(x)
```

```python
#Criando o dataframe df que armazena as características de cada espécie
df = pd.DataFrame(req_informacoes)
df.head(1)
```

```python
##Forças dos Pokemons
```

```python
#Estudando a documentação da API, stats representam as forças de cada pokemon:
  #stats(0)= HP
  #stats(1)= Ataque
  #stats(2)= Defesa
  #stats(3)= Ataque Especial 
  #stats(4)= Defesa Especial 
  #stats(5)= Velocidade

#Informações de stats coletadas ficam armazanadas no dataframe df_stats

df_stats = df['stats'].apply(pd.Series)
df_stats.head()
```

```python
#Extraindo o stats(0) e mesclando na dataframe base
df_hp=df_stats[0].apply(pd.Series)
df_hp=df_hp[['base_stat']]
df_hp = df_hp.rename(columns={'base_stat': 'HP'})

pokedex_temp1 = pd.merge(df_especies['Pokemon'], df_hp, left_index=True, right_index=True)
pokedex_temp1.head()
```

```python
##Corrente Evolutiva dos Pokemons
```

```python
#Criando loop para interar entre linhas do dataframe base df_especies
#Requisições serão armazenadas na lista req_informacoes

req_evolucao = []
for i in df['id']:
  link= "https://pokeapi.co/api/v2/pokemon-species/"+str(i)+"/"
  x = get_informacoes(link).json()
  req_evolucao.append(x)
```

```python
#Criando o dataframe df_detalhes que armazena outras características de cada espécie
df_detalhes = pd.DataFrame(req_evolucao)
df_detalhes.head()
```

```python
#Extraindo o pokemon origem e mesclando na dataframe base
df_evolucao=df_detalhes['evolves_from_species'].apply(pd.Series)
df_evolucao = df_evolucao[['name']]
df_evolucao = df_evolucao.rename(columns={'name': 'Pokemon Origem'})

pokedex_temp20 = pd.merge(pokedex_temp19, df_evolucao, left_index=True, right_index=True)
pokedex_temp20.head(3)
```

```python
#Extraindo a cor do pokemon e mesclando na dataframe base
df_cor =df_detalhes['color'].apply(pd.Series)
df_cor = df_cor[['name']]
df_cor = df_cor.rename(columns={'name': 'Cor'})

#Dataframe final
pokedex = pd.merge(pokedex_temp20, df_cor, left_index=True, right_index=True)
pokedex.head(3)
```

```python
#Salvando o dataframe final para formato CSV
pokedex.to_csv('pokedex.csv')
```
